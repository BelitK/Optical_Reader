```python
import numpy as np
import argparse
import imutils
import cv2
```

```python
image = cv2.imread(image)
```

```python
lower = np.array([0, 0, 0])
upper = np.array([45, 45, 45])
shapeMask = cv2.inRange(image, lower, upper)
```

```python
cnts = cv2.findContours(shapeMask.copy(), cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
```
